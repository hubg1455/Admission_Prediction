```python
#r^2 score
lr.score(x_test,y_test)
```

```python
#create a function to create adjusted R-Squared
def adt_r2(x,y):
    r2=lr.score(x,y)
    n=x.shape[0]
    p=x.shape[1]
    adjusted_r2=1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
adt_r2(x_test,y_test)
```
